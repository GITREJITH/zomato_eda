```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/zomato.csv")
```

```python
 df.shape
```

```python
df.head()
```

```python
columns_to_drop = ["url","address","dish_liked","phone","menu_item","reviews_list"]
df.drop(columns=columns_to_drop,inplace=True)
```

```python
df.describe()
```

```python

df.head()
```

```python
df["name"].value_counts()
```

```python
df.info()
```

dropping duplicates

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
#cleaning rate column

df["rate"].unique()
```

```python
#removing "new and "-" and /5

def handle_rate(value):
  if value == "NEW" or value == "-":
    return np.nan
  else:
    value = str(value).split("/")
    value = value[0]
    return float(value)

df["rate"] = df['rate'].apply(handle_rate)
df['rate'].head()
```

FILLING NULL VALUES IN RATE COLUMN WITH MEAN

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.rename(columns={"approx_cost(for two people)":"cost2plates",'listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
# dropping one

df = df.drop(columns=['listed_in(city)'])
```

```python
df.head()
```

```python
df['cost2plates'].unique()
```

```python
# removing ',' from cost2plates column
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

df['cost2plates'] = df['cost2plates'].apply(handlecomma)
df['cost2plates'].unique()
```

```python
df.head()
```

**cleaning rest type**

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_types(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_types)
df['rest_type'].value_counts()
```

CLEANING LOCATION COLUMN

```python
df['location'].value_counts()
```

```python
# location having less than 300 restaurant
location = df['location'].value_counts(ascending = False)
location
```

```python
location_lessthan300 = location[location<300]
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
df.shape
```

CLEANING CUISINES

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)
```

```python

cuisines_lessthan100 = cuisines[cuisines<100]
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['Type'].value_counts()
```
